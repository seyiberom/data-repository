```python
# gather daily trading data for each pair for all the tradings days we are intrsted in
# the tradings days are contained in the list "param_list"
# the trading pairs aare contained in the list "trading_list"
# we are extracting data from the binance historical data download API
# the data set for each trading pair is stord as a python dictionary

l = []
for d in param_list:
  day_data = []
  print("downloading trading data for ....  " + str(d[0]))
  pair_data = []
  for j  in trading_list:
    filename = str(j[1]) + '-1d-' + str(d[0]) + '.zip'
    url = 'https://data.binance.vision/data/spot/daily/klines/' + j[1] + '/1d/' + filename
    print(url)
    try:
        urllib.request.urlretrieve(url, filename)
        with zipfile.ZipFile(filename, 'r') as zip_ref:
          zip_ref.extractall('.')
        csvname = filename.replace("zip", "csv")
        with open(csvname,'r', encoding="utf-8") as f:
            reader = csv.reader(f)
            for row in reader:
              int_row = [float(value) for value in row]
              dd = datetime.datetime.fromtimestamp(int(row[0])/1000).strftime('%Y-%m-%d %H:%M:%S')
              int_row.insert(0, j[1])
              int_row[1] = dd

    except urllib.error.HTTPError as err:
        print(f'A HTTPError was thrown: {err.code} {err.reason}')
        # if we could not obtain data for a pair we fill it with zero value for that day
        int_row =  list(itertools.repeat(0, 13))
        int_row[0] = j[1]
        int_row[1] = d[0]
    # convert date data to readvle form
    day_data.append(int_row)
    filename = 'trade_data'+ str(d[0])+ '.csv'
  with open(filename,'w', encoding="utf-8") as f:
      writer = csv.writer(f)
      writer.writerow(['trading_pair', 'open_time', 'open', 'high', 'low', 'close', 'volume', 'close_time', 'qav', 'num_trades', 'taker_base_vol','taker_quote_vol', 'ignore'])
      for data_list in day_data:
         writer.writerow(data_list)
      f.close()



```
